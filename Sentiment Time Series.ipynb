```python
def classify_sentiment(text):
    """ Calculates the sentiment for a given text (in english) and classifies it as positive, negative or neutral"""

    from pattern.en import sentiment

    score = sentiment(text)[0]
    if score > 0:
        return 'positive'
    elif score < 0:
        return 'negative'
    else:
        return 'neutral'

```

```python
df['sentiment'] = df.text.apply(classify_sentiment)
```

```python
df['created_at'] = pd.to_datetime(df.created_at)

def append_time_codings(dataframe, column='created_at'):
    times = dataframe.apply(lambda x: pd.Series({
            'year': x[column].year,
            'month': x[column].month,
            'day': x[column].day,
            'weekday_code': x[column].weekday(),
            'weekday_label': x[column].strftime('%A')
            }), axis=1)
    return times

df = pd.concat([df, append_time_codings(df)], axis=1)
df['n'] = 1

```

```python
# sentiment
sentiment = df.groupby(['year', 'month', 'day', 'created_at', 'sentiment'])['n'].sum().reset_index()
sentiment = pd.pivot_table(sentiment, index=['year', 'month', 'day', 'created_at'], columns=['sentiment'], values='n', fill_value=0).reset_index()
```
